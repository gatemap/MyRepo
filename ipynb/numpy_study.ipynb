```python
import numpy as np      # numpy 라이브러리 가져오기

def numpy_test() :
    x = np.array([3, 1, 2]) # ndarray 타입의 배열 생성
    print(x)
    type(x)
    
    
    ndarray1 = np.array([[1, 2, 3], [4, 5, 6]])
    print(ndarray1.ndim)        # 배열의 차원 수 반환
    print(ndarray1.shape)       # 배열 각 차원의 크기를 튜플 형태로 표현
    
    print(np.dot(4, 2))         # 스칼라 곱 연산
    
    ndar1 = np.array([1, 2, 3])
    ndar2 = np.array([4, 5, 6])
    print(np.dot(ndar1, ndar2)) # 벡터 내적 연산
    
    ndar1 = np.array([[1, 2], [3, 4]])
    ndar2 = np.array([[1, 2], [3, 4]])
    print(np.dot(ndar1, ndar2))     # 행렬곱 연산
    
    print(np.matmul(ndar1, ndar2))  # 행렬의 행렬곱 연산? 외적?
    
    ndar1 = np.array([[1, 2, 3], [4, 5, 6]])
    ndar2 = np.array([[6, 3], [5, 2], [4, 1]])
    print(np.matmul(ndar1, ndar2))
    
    ndar1 = np.array([1, 2, 3])
    print(np.matmul(ndar1, ndar2))
    
    # m*n 크기의 영행렬(값이 모두 0) 생성
    print(np.zeros((2, 2)))
    
    # m*n 크기의 유닛행렬(값이 모두 1) 생성
    print(np.ones(2, 3))
    
    # m*n 크기의 모든 원소가 k인 행렬 생성
    print(np.full((2, 3), 5))
    
    # n*n 크기의 단위 행렬 생성
    print(np.eye(3))
    
    # 행렬의 평탄화 작업. 평탄화 작업 : N차원 배열을 1차원 배열로 변환시켜주는 작업
    ndarray2 = np.array([[6, 3], [5, 2], [4, 1]])
    print(ndarray2.flatten())
    
    # 배열의 차원을 변경
    print(ndarray2.shape)
    print(ndarray2.reshape(2, 3))
    
    # 특정 배열을 만들어줌. .arange([start, ], stop, [step, ])
    print(np.arange(5))
    print(np.arange(5, 10))
    print(np.arange(5, 10, 2))
    
    # random .random.randint(low, [high, ] [size, ])
    print(np.random.randint(5))     # 0~5까지의 난수
    print(np.random.randint(5, 10)) # 5~9까지의 난수
    print(np.random.randint(5, size=(2, 4)))
    
    # .random.rand(m, n) : m*n 크기의 배열 생성 및 0~1 사이의 난수로 초기화
    print(np.random.rand(5))
    print(np.random.rand(2, 3))
    
    # numpy 조건식
    ndarray3 = np.array([[6, 3], [5, 2], [4, 1]])
    print(ndarray3 > 4)
    print(ndarray3[ndarray3 > 4])
    
    
def numpy_practice() : 
    # 1
    ndarr = np.zeros((10))
    print(ndarr)
    
    # 2
    ndarr[4] = 1
    print(ndarr)
    
    # 3
    print(np.arange(10, 31))
    
    # 4
    print(np.random.randint(99, size=(2, 2)))
    
    # 5 
    print(np.random.rand(2, 4))
    
    # 6
    ndarr = np.arange(35, 75)
    ndarr = ndarr.reshape(10, 4)
    print(ndarr)
    
    # 7
    ndarr2 = np.flip(ndarr)
    print(ndarr2)
    
    # 8
    print(ndarr2[1:-1, 2:])
    
    # 9
    #ndarr3 = ndarr2[:, -1:].reshape(1, 4)
    #print(ndarr3)
    
    # 9-1
    #print(np.array([ndarr4[]]))
    
    
def main() :
    #practice_1()
    #numpy_test()
    numpy_practice()
    
    
if __name__ == '__main__' :
    main()
```
